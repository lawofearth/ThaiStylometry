# Work code

```python
import os, sys
from pandas import DataFrame as df
import pandas as pd
import numpy as np
import random

data = pd.read_csv("Input_StylometricFeatures_190314_for_real.csv", header=None)

new_data = data.copy()
author = 0
book = 0
for i in range(len(data)):
    # initial
    if author == 0 and book == 0:
        author+=1
        book = author * 1000
        new_data.loc[i, 0] = book
        new_data.loc[i, 1] = author
        
    elif data.loc[i,1] != data.loc[i-1,1]:
        author+=1
        book = author * 1000
        new_data.loc[i, 0] = book
        new_data.loc[i, 1] = author
        
    elif data.loc[i,0] != data.loc[i-1,0]:
        book += 1
        new_data.loc[i, 0] = book
        new_data.loc[i, 1] = author
    else:
        new_data.loc[i, 0] = book
        new_data.loc[i, 1] = author
        
data = new_data     

author_table = df()
author = 0
book = 0
for i in range(len(data)):
    # initial
    if author == 0 and book == 0:
        author_table.loc[author, 0] = data.loc[i, 1]
        book += 1
        author_table.loc[author, book] = data.loc[i, 0]
        
    elif data.loc[i,1] != data.loc[i-1,1]:
        author+=1
        author_table.loc[author, 0] = data.loc[i, 1]
        book = 1
        author_table.loc[author, book] = data.loc[i, 0]
        
    elif data.loc[i,0] != data.loc[i-1,0]:
        book += 1
        author_table.loc[author, book] = data.loc[i, 0]
        
```

```python
def export_queryset(k, author_table):
    random_matrix = df()
    openset_matrix = df()
    for i in range(len(author_table)):
        # random book each author
        random_matrix.loc[i,0] = author_table.loc[i,int(np.random.uniform(0,sum(~author_table.loc[i,:].isna())-1))+1]
       
    # for random authors   
    openset_input = random_matrix.sample(frac=0.5, random_state=k)
    # for openset
    authorset_inlist = author_table.loc[openset_input.index.tolist(),0]
    
    random_matrix.to_csv("0%.0f.csv" %k, header=False, index = False)
    openset_input.to_csv("openset0%.0f.csv" %k, header=False, index = False)
    authorset_inlist.to_csv("openset_authorlist0%0.f.csv" %k, header=False, index = False)
    
```

```python
for j in range(1,6):
    export_queryset(j, author_table)
```
