<h1>ลบช่องว่างออก

```python
def clean_document(text): #ลบช่องว่าง
    text = "".join(text.split())
    return text
```

<h1> ตัด 1002 character และตัด n grams

```python
def word2ngrams(text, n_grams,word_in_doc): #ตัดตัวอักษร n grams    
    #token character
    count = 0
    word_in_doc=[]
    char = ''
    for i in text:
        if count <= 1001:
            char+=i
            count+=1
        else:
            word_in_doc.append(char)
            char = ''
            char+=i
            count=0
            
    n_grams_word = []
    for word in word_in_doc:      
        word_grams = ["".join(j) for j in zip(*[word[i:] for i in range(n_grams)])]
        n_grams_word.append(word_grams)
    
    return n_grams_word,word_in_doc
```

<h1>นับจำนวนคำทั้งหมด

```python
def count_ngrams(doc_token_word): #นับจำนวน n grams
    ngram_character_freq_list = []
    
    for ngrams_output in doc_token_word[0]:
        characte_ngrams = list(set(ngrams_output)) #เอาคำซ้ำออก
        count_word_dup=[] #นับจำนวนคำ
        
        for w in characte_ngrams:
            count_word_dup.append(ngrams_output.count(w))
            
        ngram_character_freq = dict(zip(characte_ngrams, count_word_dup))
        #print(ngram_character_freq)
        ngram_character_freq_list.append(ngram_character_freq)
        
    return ngram_character_freq_list
```

<h1> เรียงฟังก์ชั่นนี้อันเดียว ก็จะเรียกทุกอันเอง

```python
def graph_ngram(max_list): #เอาไว้ดูความถี่ของคำ
    for doc in max_list:
        for sub_doc in doc:
            y_value = []
            x_items = []
            for y in sub_doc:
                y_value.append(y[1])
                x_items.append(y[0])
            #print('end 10 sentence')  
            x_len = range(len(y_value))
            #print('word = {} and value = {} '.format(x_items,y_value))
            fig,ax = plt.subplots()
            plt.bar(x_len, y_value , align='center', alpha=0.5)
            plt.xticks(x_len, x_items,fontname='Tahoma')
            plt.ylabel('Frequency')
            plt.title('TF-IDF Frequency')
            plt.grid()
            fig.autofmt_xdate()
            plt.show()
```

```python
def max_min_index(tf_doc): #หา max min
    max_list = []
    for doc in tf_doc:
        max_sub_list = []
        for sub_doc in doc:
            word = list(sub_doc.keys())
            value = list(sub_doc.values())
            max_items = []
            while len(max_items) < 500:
                max_index = word.index(max(word))
                #print('index {} items {} value {}'.format(max_index,characte_ngrams[max_index],count_word_dup[max_index]))
                if(value[max_index] != 0):
                    max_items.append([word[max_index],value[max_index]])
                del word[max_index]
                del value[max_index]
            max_sub_list.append(max_items)
        max_list.append(max_sub_list)
    return max_list
```

```python
def edit_distance(list_A,list_B):
    A_items = []
    A_values = []
    B_items = []
    B_values = []
    for i in list_A:
        A_items.append(i[0])
        A_values.append(i[1])
    for i in list_B:
        B_items.append(i[0])
        B_values.append(i[1])
    matrixAB = []
    for a in A_items:
        sub_matrixAB = []
        for b in B_items:
            sub_matrixAB.append(editdistance.eval(a,b))
        matrixAB.append(sub_matrixAB)
    return A_values,B_values,matrixAB
```
