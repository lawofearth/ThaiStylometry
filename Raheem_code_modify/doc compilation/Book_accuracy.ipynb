## code -> entropy level, top 50 pct, average

```python
doc_frag = []
doc_frag = pd.read_csv("frag_author_percentage_F_5_K_6_Q_05.csv")

book_set = set(doc_frag.loc[:,'book'].dropna())

doc_frag_pure = doc_frag.drop(columns = ["Unnamed: 0","book","author"])
doc_count = 0
correct_doc = 0 


for book_i in book_set:
    
    total_prob_book = doc_frag_pure.loc[doc_frag.loc[:,'book'] == book_i,:].sum()
    
    if total_prob_book.idxmax() != '0.0':
        doc_count += 1
        
        author_list = doc_frag.loc[doc_frag.loc[:,'book'] == book_i,'author']
    
        book_frag_pure = doc_frag_pure.loc[doc_frag.loc[:,'book'] == book_i,:]
        # entropy is -sum(p*lnp)
        doc_frag_pure_entropy = book_frag_pure.applymap(lambda x: -x*np.log(x)).sum(axis = 1)
    
        p50 = int(len(doc_frag_pure_entropy)/2)
        idx_min = doc_frag_pure_entropy.idxmin()
        total_prob_book = book_frag_pure.loc[idx_min,:]
        doc_frag_pure_entropy = doc_frag_pure_entropy.drop(idx_min)

        for frag_prob in range(1,p50):
            # choose p50 minimum entropy 
            idx_min = doc_frag_pure_entropy.idxmin()
            total_prob_book += book_frag_pure.loc[idx_min,:]
            doc_frag_pure_entropy = doc_frag_pure_entropy.drop(idx_min)
    
    
    if total_prob_book.idxmax() == str(author_list.iloc[0]):
        # check results
        correct_doc += 1
        
float(correct_doc/doc_count)       
```
